```python
#EXTRACT  (Types available: informal, oficial)
def extract_from_api(dolar_type):
  return requests.get(f'https://mercados.ambito.com//dolar/{dolar_type}/variacion').json()

respuesta =[]

for dolar in["informal", "oficial","dolarqatar"]:  ##sume a mano el dolarqatar
  info =extract_from_api(dolar)
  info["tipo_dolar"]=dolar
  respuesta.append(info)
```

```python
pd.DataFrame(respuesta)
```

```python
dolar = pd.DataFrame(respuesta)
```
